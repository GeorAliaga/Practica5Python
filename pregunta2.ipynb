```python
import pandas as pd

df_wine = pd.read_csv("data/winemag-data-130k-v2.csv")
df_wine.head()
```

```python
import pandas as pd

df_wine.rename(columns={'country': 'pais', 'variety': 'variedad', 'winery': 'bodega'}, inplace=True)

df_wine['precio_usd'] = df_wine['price'] * 1.18  
df_wine['rango_puntuacion'] = pd.cut(df_wine['points'], bins=[80, 85, 90, 95, 100], labels=['80-85', '86-90', '91-95', '96-100'])
df_wine['valoracion'] = df_wine['points'] / df_wine['price']  


reporte1 = df_wine.groupby('pais')['points'].mean().sort_values(ascending=False).head(10)
reporte2 = df_wine.groupby(['pais', 'rango_puntuacion'])['title'].count().unstack()
reporte3 = df_wine.groupby('variedad')['precio_usd'].mean().sort_values(ascending=False).head(10)
reporte4 = df_wine.groupby('bodega')['valoracion'].mean().sort_values(ascending=False).head(10)


reporte1.to_excel('reporte_puntuacion_por_pais.xlsx', index=True)


print("Renombres y creacion de columnas realizados:")
print(df_wine.head())

print("\nReporte 1: Puntuacion promedio por pais (Top 10):")
print(reporte1)

print("\nReporte 2: Cantidad de vinos en cada rango de puntuacion por pais:")
print(reporte2)

print("\nReporte 3: Precio promedio por variedad (Top 10):")
print(reporte3)

print("\nReporte 4: Valoración promedio por bodega (Top 10):")
print(reporte4)
```
